```python
#Step1:importing the dataset.
import pandas as pd
df=pd.read_csv(r"C:\Users\Dell\Desktop\D.S Assignments\Basic stats - 1\sales_data_with_discounts.csv")
df
```

```python
#step2: identifing the numerical columns
num_columns=df.select_dtypes(include=['int64','float64'])
num_columns
```

```python
#step3: basic statictics
import matplotlib.pyplot as plt
num_columns.mean()
```

```python
num_columns.median()
```

```python
num_columns.mode()
```

```python
num_columns.std()
```

```python
num_columns.describe()
```

```python
#step2:To visualize the distribution and relationship of numerical and
#categorical variables in the dataset.
plt.hist(num_columns,bins=20,color=["blue","red","black","orange","pink","green"],edgecolor="white",rwidth=1)
plt.title("Histogram of numerical columns")
plt.show()
```

```python
num_columns.skew()
```

```python
num_columns.plot(kind='box', subplots=True, layout=(2, 3), figsize=(15, 10), sharex=False, sharey=False, patch_artist=True)
plt.show()
```

```python
q1=num_columns.quantile(0.25)
q3=num_columns.quantile(0.75)
IQR=q3-q1
print("Interquartile Range (IQR) for each numerical column:\n", IQR)
```

```python
catego_columns=df.select_dtypes(include=['object','category'])
catego_columns
```

```python
for col in catego_columns:
    plt.figure(figsize=(10, 5))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'Frequency of Categories in {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
#Discuss the need for converting categorical data into dummy variables 
#1.It helps work efficient in numerical data than categorical data
# one hot encoding helps categorical data into binary number( 0 or 1)
```

```python
#selecting categorical columns only
cat_columns=df.select_dtypes(include=['object'])
cat_columns
```

```python
#apply one hot encoding
cat_columns_encoded = pd.get_dummies(cat_columns, columns=['Day','SKU','City','BU','Brand','Model'])

# Display the transformed DataFrame
print("\nTransformed DataFrame (after one-hot encoding):")
cat_columns_encoded
```
